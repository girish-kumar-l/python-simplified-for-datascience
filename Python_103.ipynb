# **Mathplotlib**


1. This is a basic graph plotting library, mostly developed in python only
2. pyplot is one important submodule where most of the commonly used function are available
3. 

```python
#basic plotting code. Drawing a line from x1,y1 to x2,y2
import matplotlib.pyplot as plt
import numpy as np

xpoints = [10, 70]
ypoints = [0, 100]

plt.plot(xpoints, ypoints)
plt.show()
```

```python
#only points can also be shown
plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
#without passing x values, we default them with 0,1,2, etc as long as we have some y values

ypoints = [20,5,25,50,20,10]

plt.plot(ypoints)
plt.show()
```

```python
#passing fmt strings can help in better formating. fmt strings are basically "marker|line|color"

plt.plot(ypoints, "*:r") # * to mark a point, : means dotted lines, r means red color lines
plt.show()
```

```python
#plot() function allows you to draw as many lines as you want.
ypoints1 = [20,5,25,50,20,10]
ypoints2 = [30,50,40,10,20,30]

plt.plot(ypoints1)
plt.plot(ypoints2)
```

```python
#creating titles and labels for axis can be done using xlabel() function
plt.plot(ypoints1)
plt.plot(ypoints2)
plt.title("Sample plotting exercise")
plt.xlabel("my x-axis")
plt.ylabel("my y-axis")
plt.show()

#Note: Additional capabilites like font, styles are also offered in the function. Maybe useful for visualization. ommiting since the focus is on deriving insights from data
```

```python
#subplot function allows users to create multiple plots using multiple data points. subplot() function consumes 3 parameters (row, col, position)
#plot 1:

#source: w3schools.com
x = [0, 1, 2, 3]
y = [3, 8, 1, 10]

plt.subplot(2, 1, 1) #2rows, 1 column, 1st position to be taken
plt.plot(x,y)

#plot 2:
x = [0, 1, 2, 3]
y = [10, 20, 30, 40]

plt.subplot(2, 1, 2) #2rows, 1 column, 2nd position to be taken
plt.plot(x,y)

plt.show()
```

```python
#scatter plot and histogram
x = [0, 1, 2, 3]
y = [10, 20, 30, 40]
plt.subplot(1, 2, 1)
plt.scatter(x,y)

x = ["Apple","Banana","Cherry"]
y = [10, 20, 30]
plt.subplot(1, 2, 2)
plt.bar(x,y)
plt.show()
```

```python
#pie charts with labeling and explode

x = [10, 35, 20,10,25]
my_label = ["Apple","Banana","Cherries","Dates", "Mango"]
plt.subplot(1,2,1)
plt.pie(x, labels = my_label)

x = [10, 35, 20,10,25]
my_label = ["Apple","Banana","Cherries","Dates", "Mango"]
myexplode = [0.2, 0, 0, 0,0] #lets highlight apple sales by exploding it separately
plt.subplot(1,2,2)
plt.pie(x, labels = my_label,explode = myexplode)
plt.show()
```
